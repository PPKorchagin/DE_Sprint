## 2. Построение графика на по функции

```python
x = np.arange(-2,2,0.0001)

f1= lambda x: ((x**2)**(1/3)+np.sqrt((x**4)**1/3-4*(x**2-1)))/2
f2= lambda x: ((x**2)**(1/3)-np.sqrt((x**4)**1/3-4*(x**2-1)))/2

plt.plot(x,f1(x),'r')
plt.plot(x,f2(x),'r')
plt.grid()
plt.show()
```

### Построение графиков

```python
Labels = ['Age','Sex','BMI','Children','Smoker','Region','Expenses']
Vars = [age,sex,bmi,children,smoker,region,expenses]
numVars = [age,bmi,children,expenses]
numLabels = ['Age','BMI','Children','Expenses']
```

```python
n=len(numVars)
for i in range(n):
  for j in range(i,n):
    if i != j:
      x,y,xLab,yLab=numVars[i],numVars[j],numLabels[i],numLabels[j]
      plt.scatter(x[sex==1],y[sex==1],marker=".",c='blue',label='male')
      plt.scatter(x[sex==0],y[sex==0],marker=".",c='red',label='female')
      plt.xlabel(xLab)
      plt.ylabel(yLab)
      plt.title(yLab+ '(' +xLab+')')
      plt.legend()
      plt.show()
      
      plt.scatter(x[smoker==1],y[smoker==1],marker=".",c='black',label='smoker')
      plt.scatter(x[smoker==0],y[smoker==0],marker=".",c='yellow',label='no smoker')
      plt.xlabel(xLab)
      plt.ylabel(yLab)
      plt.title(yLab+ '(' +xLab+')')
      plt.legend()
      plt.show()

      plt.scatter(x[region==0],y[region==0],marker=".",label='northwest')
      plt.scatter(x[region==1],y[region==1],marker=".",label='northeast')
      plt.scatter(x[region==2],y[region==2],marker=".",label='southwest')
      plt.scatter(x[region==3],y[region==3],marker=".",label='southeast')
      plt.xlabel(xLab)
      plt.ylabel(yLab)
      plt.title(yLab+ '(' +xLab+')')
      plt.legend()
      plt.show()
```
