```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(-2,2,0.0001)

f1= lambda x: ((x**2)**(1/3)+np.sqrt((x**4)**1/3-4*(x**2-1)))/2
f2= lambda x: ((x**2)**(1/3)-np.sqrt((x**4)**1/3-4*(x**2-1)))/2

plt.plot(x,f1(x),'r')
plt.plot(x,f2(x),'r')
plt.grid()
plt.show()
```
